# Step 2 — Choose the correct ones for your analysis

```python
# Step 2: Clean and combine key FRA datasets
import pandas as pd
from pathlib import Path

# Paths
project_root = Path(r"C:\Users\Administrator\Desktop\Data Analytics\Portfolio project\Global-Forest-Trends")
raw_dir = project_root / "data" / "raw" / "FRA_Years_variables"
processed_dir = project_root / "data" / "processed"
processed_dir.mkdir(parents=True, exist_ok=True)

# Helper: tidy one FRA CSV into long format
def tidy_fra_csv(path: Path, variable_type: str) -> pd.DataFrame:
    """
    Reads a wide-format FRA CSV, selects key columns,
    melts years into long format, and returns a tidy DataFrame.

    Parameters
    ----------
    path : Path
        Path to the CSV file.
    variable_type : str
        Short name identifying the dataset (e.g., 'forest_area').

    Returns
    -------
    pd.DataFrame
        Columns: region, iso3, country, year, value, variable_type
    """
    df = pd.read_csv(path)

    # rename for clarity
    df = df.rename(columns={
        "regions": "region",
        "name": "country"
    })

    # keep only region, iso3, country + years
    year_cols = [c for c in df.columns if c.isdigit()]
    keep_cols = ["region", "iso3", "country"] + year_cols
    df = df[keep_cols]

    # melt years to long format
    df_long = df.melt(
        id_vars=["region", "iso3", "country"],
        value_vars=year_cols,
        var_name="year",
        value_name="forest_area_kha"
    )

    # add variable tag
    df_long["variable_type"] = variable_type
    return df_long

# Files to use
files = {
    "forest_area": "1a_forestArea_2025_10_22.csv",
    "nat_regen_forest": "1b_naturallyRegeneratingForest_2025_10_22.csv",
    "planted_forest": "1b_otherPlantedForest_2025_10_22.csv"
}

# Combine all three
frames = []
for var_type, fname in files.items():
    path = raw_dir / fname
    tidy_df = tidy_fra_csv(path, var_type)
    frames.append(tidy_df)
    print(f"✓ Tidied {var_type}: {len(tidy_df)} rows")

combined = pd.concat(frames, ignore_index=True)

# Ensure numeric types
combined["year"] = combined["year"].astype(int)
combined["forest_area_kha"] = pd.to_numeric(combined["forest_area_kha"], errors="coerce")

# Save
out_path = processed_dir / "forest_area_tidy.csv"
combined.to_csv(out_path, index=False)
print(f"\n💾 Saved tidy dataset → {out_path}")
display(combined.head())
```

# Step 3 — Calculating Change Metrics (1990 → 2025)

```python
# Step 3: Calculate forest change metrics 1990 → 2025
import pandas as pd
from pathlib import Path

project_root = Path(r"C:\Users\Administrator\Desktop\Data Analytics\Portfolio project\Global-Forest-Trends")
processed_dir = project_root / "data" / "processed"
outputs_dir = project_root / "outputs"
outputs_dir.mkdir(parents=True, exist_ok=True)

# Load tidy dataset
df = pd.read_csv(processed_dir / "forest_area_tidy.csv")

# Filter only total forest area for the main analysis
forest_df = df[df["variable_type"] == "forest_area"].copy()

# Pivot to wide: one row per country, columns for each year
wide = forest_df.pivot_table(
    index=["region", "iso3", "country"],
    columns="year",
    values="forest_area_kha"
).reset_index()

# Compute changes
wide["abs_change_kha"] = wide[2025] - wide[1990]
wide["pct_change"] = ((wide[2025] - wide[1990]) / wide[1990]) * 100
wide["CAGR_%"] = ((wide[2025] / wide[1990]) ** (1/35) - 1) * 100

# Sort and create top 10 tables
top_losses = wide.sort_values("abs_change_kha").head(10)
top_gains = wide.sort_values("abs_change_kha", ascending=False).head(10)

# Save outputs
top_losses.to_csv(outputs_dir / "top10_losses.csv", index=False)
top_gains.to_csv(outputs_dir / "top10_gains.csv", index=False)

print("💾 Saved:")
print("-", outputs_dir / "top10_losses.csv")
print("-", outputs_dir / "top10_gains.csv")

display(wide.head())
```

# Step 4 – Visualization

```python
# Step 4.0: Setup (paths + load data + modern styling)
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- Paths ---
project_root = Path(r"C:\Users\Administrator\Desktop\Data Analytics\Portfolio project\Global-Forest-Trends")
processed_dir = project_root / "data" / "processed"
outputs_dir = project_root / "outputs"
figs_dir = outputs_dir / "figures"
figs_dir.mkdir(parents=True, exist_ok=True)

# --- Data ---
df = pd.read_csv(processed_dir / "forest_area_tidy.csv")
forest = df[df["variable_type"] == "forest_area"].copy()

# --- Plot style (modern + colorblind friendly) ---
sns.set_theme(style="whitegrid")                    # seaborn look
sns.set_palette("Dark2")                            # colorful but tasteful
plt.rcParams.update({
    "figure.figsize": (9, 5),
    "axes.titlesize": 14,
    "axes.labelsize": 12,
    "axes.edgecolor": "#334",
    "axes.linewidth": 1.0,
    "font.family": "sans-serif",
})

# --- Helper: save AND show ---
def save_and_show(path: Path | None):
    """
    Save the current Matplotlib figure to 'path' (if provided) and also display it inline.
    """
    if path is not None:
        plt.savefig(path, bbox_inches="tight", dpi=160)
        print(f"💾 Saved figure → {path}")
    plt.show()  # always show in the notebook
```

## 4.1 Global forest area trend (1990 → 2025)

```python
 # Step 4.1: Global forest area trend (1990 → 2025)
"""
Purpose:
    Visualize the total global forest area (in thousand hectares)
    across all reporting countries between 1990 and 2025.

Why:
    - Answers: “How has global forest area changed from 1990 → 2025?”
    - Provides context for regional and country-level analyses.

Output:
    1) Saved PNG → outputs/figures/global_forest_trend.png
    2) Inline chart in the notebook
"""

import seaborn as sns
from pathlib import Path

# --- group to global totals ---
g_global = forest.groupby("year", as_index=False)["forest_area_kha"].sum()

# --- seaborn style (modern) ---
sns.set_theme(style="whitegrid", palette="Dark2")

# --- helper: save then show (avoid blank-figure issue) ---
def save_then_show(path: Path):
    """Save the current figure first (prevents blank files), then display inline."""
    plt.savefig(path, bbox_inches="tight", dpi=160)  # save BEFORE show
    print(f"💾 Saved figure → {path}")
    plt.show()                                       # show after saving

# --- plot ---
fig, ax = plt.subplots(figsize=(9, 5))
sns.lineplot(data=g_global, x="year", y="forest_area_kha", marker="o", linewidth=2.5, ax=ax)

ax.set_title("🌍 Global Forest Area, 1990–2025", fontsize=14, pad=10)
ax.set_xlabel("Year", fontsize=12)
ax.set_ylabel("Forest Area (Thousand Hectares)", fontsize=12)
ax.grid(True, linestyle="--", alpha=0.5)

# --- save first, then show ---
save_then_show(figs_dir / "global_forest_trend.png")
plt.close(fig)  # tidy up
```

## 4.2 Regional trends (one figure per region)

```python
# Step 4.2: Regional trends (one figure per region)
"""
Purpose
-------
Plot a separate time-series chart for each FAO region showing the sum of
forest area (kha) across countries from 1990–2025.

Why
---
This answers: “Which regions gained or lost the most forest area?”
It provides context before looking at country-level extremes.

Output
------
For each region:
  1) PNG saved to: outputs/figures/region_trend_<region>.png
  2) Inline chart displayed in the notebook

Notes
-----
- We aggregate to regional totals per year (sum of countries).
- We 'save THEN show' to avoid blank images on some backends.
- Filenames are sanitized to avoid issues with spaces/slashes.
"""

import seaborn as sns
from pathlib import Path

# --- ensure a modern, clean look (okay to call again) ---
sns.set_theme(style="whitegrid", palette="Dark2")

# --- helper: save first, then show (avoid blank files) ---
def save_then_show(path: Path):
    """Save current figure to 'path' (PNG), then display inline."""
    plt.savefig(path, bbox_inches="tight", dpi=160)  # save BEFORE show
    print(f"💾 Saved figure → {path}")
    plt.show()

def safe_name(s: str) -> str:
    """Make a region name safe for filenames (lowercase, underscores, no slashes)."""
    return s.lower().replace(" ", "_").replace("/", "-")

# --- loop each region and plot ---
for region, g in forest.groupby("region"):
    # Aggregate to regional total per year and sort by year
    g_region = (
        g.groupby("year", as_index=False)["forest_area_kha"]
         .sum()
         .sort_values("year")
    )

    # Create figure
    fig, ax = plt.subplots(figsize=(9, 5))
    sns.lineplot(
        data=g_region,
        x="year", y="forest_area_kha",
        marker="o", linewidth=2.5, ax=ax
    )

    # Titles/labels
    ax.set_title(f"{region}: Forest Area, 1990–2025 (kha)", fontsize=14, pad=10)
    ax.set_xlabel("Year", fontsize=12)
    ax.set_ylabel("Forest Area (Thousand Hectares)", fontsize=12)
    ax.grid(True, linestyle="--", alpha=0.5)

    # Save then show (prevents blank files), then tidy up
    fname = figs_dir / f"region_trend_{safe_name(region)}.png"
    save_then_show(fname)
    plt.close(fig)
```

## 4.3 Top 10 loss / gain bars (1990 → 2025)

```python
 # -------------------------------------------------------------------
# ➋ TOP 10 GAIN COUNTRIES (1990 → 2025)
# -------------------------------------------------------------------
"""
We sort by 'abs_change_kha' ascending = smaller on top, larger (gains) at bottom.
Bars use green to emphasize reforestation success.
"""
gains_plot = top_gains.sort_values("abs_change_kha", ascending=True)

fig, ax = plt.subplots(figsize=(9, 5))   # ✅ fixed here (was plt.subpl)
sns.barplot(
    data=gains_plot,
    y="country", x="abs_change_kha",
    color="#5CB85C", ax=ax
)

ax.set_title("🌱 Top 10 Countries – Forest Area Gain (1990 → 2025)", fontsize=14, pad=10)
ax.set_xlabel("Absolute Change (Thousand Hectares)", fontsize=12)
ax.set_ylabel("")
ax.grid(True, linestyle="--", alpha=0.4)

save_then_show(figs_dir / "top10_gains_bar.png")
plt.close(fig)
```

## 4.4 (Optional) Natural vs Planted over time — if you want a comparison

```python
# Step 4.4 (Optional): Natural vs Other Planted Forests (Global Totals)
"""
Purpose
-------
Compare how the total area of *naturally regenerating* forests and *other planted*
forests evolved globally from 1990 → 2025.

Why
---
This visualization highlights whether reforestation is balancing natural forest loss,
and how each type contributes to total forest cover.

Caveat
------
“Other planted” represents only one planted-forest subtype; it is *not* the total planted forest category.

Output
------
✅ Inline chart (Seaborn lineplot)
💾 PNG saved to: outputs/figures/global_natural_vs_other_planted.png
"""

import seaborn as sns
from pathlib import Path

# --- Filter the relevant subsets ---
nat = df[df["variable_type"] == "nat_regen_forest"]
opl = df[df["variable_type"] == "planted_forest"]  # 'otherPlantedForest'

# --- Aggregate by year (global totals) ---
g_nat = nat.groupby("year", as_index=False)["forest_area_kha"].sum()
g_opl = opl.groupby("year", as_index=False)["forest_area_kha"].sum()

# --- Combine into one tidy frame for seaborn ---
g_nat["type"] = "Naturally Regenerating"
g_opl["type"] = "Other Planted"
g_all = pd.concat([g_nat, g_opl], ignore_index=True)

# --- Styling ---
sns.set_theme(style="whitegrid", palette="Set2")

# --- Plot ---
fig, ax = plt.subplots(figsize=(9, 5))
sns.lineplot(
    data=g_all,
    x="year", y="forest_area_kha",
    hue="type", marker="o", linewidth=2.5, ax=ax
)

ax.set_title("🌍 Global: Natural vs Other Planted Forest (1990–2025)", fontsize=14, pad=10)
ax.set_xlabel("Year", fontsize=12)
ax.set_ylabel("Forest Area (Thousand Hectares)", fontsize=12)
ax.grid(True, linestyle="--", alpha=0.5)
ax.legend(title="Forest Type", loc="best")

# --- Save then show ---
save_then_show(figs_dir / "global_natural_vs_other_planted.png")
plt.close(fig)
```

# # Step 6: Kenya Forest Trends (1990–2025)

```python
 # Step 6: Kenya Forest Trends (1990–2025)
"""
Purpose
-------
Create a Kenya-focused mini-analysis:
  • Trend of Kenya's forest area (1990–2025)
  • Comparison vs Africa and the World
  • (Optional) Forest area as % of Kenya's land area

Why
---
Adds local context for a Kenya case study:
  • Are Kenya's forests growing or shrinking?
  • How does Kenya compare to the continental/global trend?
  • Is Kenya moving toward national forest-cover targets?

Outputs
-------
Inline charts (Seaborn) + PNG copies saved to:
  - outputs/figures/kenya_forest_trend.png
  - outputs/figures/kenya_africa_world_comparison.png
  - outputs/figures/kenya_forest_pct_land.png  (optional)
And printed metrics (absolute change, % change, CAGR).
"""

from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- Ensure helper exists: save FIRST, then show (prevents blank files) ---
def save_then_show(path: Path):
    """Save current figure to 'path' (PNG), then display inline."""
    plt.savefig(path, bbox_inches="tight", dpi=160)
    print(f"💾 Saved figure → {path}")
    plt.show()

# --- Seaborn theme (consistent with earlier steps) ---
sns.set_theme(style="whitegrid", palette="Dark2")

# --- Filter Kenya (use total forest area) ---
kenya = forest[forest["iso3"] == "KEN"].sort_values("year").copy()

# --- Safety check: ensure expected years exist ---
expected_years = {1990, 2000, 2010, 2015, 2020, 2025}
missing = expected_years - set(kenya["year"].unique())
if missing:
    print(f"⚠️ Kenya missing years: {sorted(missing)} (continuing with available data)")

# --- Compute change metrics (1990 → 2025) ---
base = float(kenya.loc[kenya["year"] == 1990, "forest_area_kha"].iloc[0])
latest = float(kenya.loc[kenya["year"] == 2025, "forest_area_kha"].iloc[0])
abs_change = latest - base
pct_change = (abs_change / base) * 100
years_between = 2025 - 1990
cagr = ((latest / base) ** (1 / years_between) - 1) * 100

print("🇰🇪 Kenya — Forest Area Change (1990 → 2025)")
print(f"  1990:  {base:,.0f} kha")
print(f"  2025:  {latest:,.0f} kha")
print(f"  Δ Abs:  {abs_change:,.1f} kha")
print(f"  Δ %:    {pct_change:.2f}%")
print(f"  CAGR:   {cagr:.3f}% per year\n")

# --- 6.1 Kenya trend line ---
fig, ax = plt.subplots(figsize=(9, 5))
sns.lineplot(data=kenya, x="year", y="forest_area_kha", marker="o", linewidth=3, ax=ax)
ax.set_title("🇰🇪 Kenya — Forest Area Trend (1990–2025)", fontsize=14, pad=10)
ax.set_xlabel("Year", fontsize=12)
ax.set_ylabel("Forest Area (Thousand Hectares)", fontsize=12)
ax.grid(True, linestyle="--", alpha=0.5)
save_then_show(figs_dir / "kenya_forest_trend.png")
plt.close(fig)

# --- 6.2 Kenya vs Africa vs World ---
world = df.groupby("year", as_index=False)["forest_area_kha"].sum()
africa = df[df["region"] == "Africa"].groupby("year", as_index=False)["forest_area_kha"].sum()

fig, ax = plt.subplots(figsize=(9, 5))
sns.lineplot(data=world,  x="year", y="forest_area_kha", marker="o", linewidth=2.5, label="World",  ax=ax)
sns.lineplot(data=africa, x="year", y="forest_area_kha", marker="o", linewidth=2.5, label="Africa", ax=ax)
sns.lineplot(data=kenya,  x="year", y="forest_area_kha", marker="o", linewidth=3.0,  label="Kenya",  ax=ax)

ax.set_title("Forest Area — Kenya vs Africa vs World (1990–2025)", fontsize=14, pad=10)
ax.set_xlabel("Year", fontsize=12)
ax.set_ylabel("Forest Area (Thousand Hectares)", fontsize=12)
ax.legend(title="Series")
ax.grid(True, linestyle="--", alpha=0.5)
save_then_show(figs_dir / "kenya_africa_world_comparison.png")
plt.close(fig)

# --- 6.3 (Optional) Forest area as % of land area for Kenya ---
"""
We compute:
    forest_%_of_land(year) = forest_area_kha(year) / land_area_kha * 100
Land area is pulled from '1a_landArea_*.csv' in data/raw/FRA_Years_variables.
"""

try:
    raw_dir = project_root / "data" / "raw" / "FRA_Years_variables"
    land_file = next(raw_dir.glob("1a_landArea_*.csv"))  # pick the file regardless of export date
    land_df = pd.read_csv(land_file)
    land_df = land_df.rename(columns={"iso3": "iso3", "name": "country"})
    kenya_land = float(land_df.loc[land_df["iso3"] == "KEN", "2025"].iloc[0])  # FRA land area is constant per country

    kenya_pct = kenya.copy()
    kenya_pct["forest_pct_land"] = (kenya_pct["forest_area_kha"] / kenya_land) * 100

    fig, ax = plt.subplots(figsize=(9, 5))
    sns.lineplot(data=kenya_pct, x="year", y="forest_pct_land", marker="o", linewidth=3, color="#2E8B57", ax=ax)
    ax.set_title("🇰🇪 Kenya — Forest Area as % of Land (1990–2025)", fontsize=14, pad=10)
    ax.set_xlabel("Year", fontsize=12)
    ax.set_ylabel("Forest Area (% of Land)", fontsize=12)
    ax.grid(True, linestyle="--", alpha=0.5)
    save_then_show(figs_dir / "kenya_forest_pct_land.png")
    plt.close(fig)
except StopIteration:
    print("ℹ️ Skipped % of land chart: no '1a_landArea_*.csv' found in data/raw/FRA_Years_variables/")
except Exception as e:
    print(f"ℹ️ Skipped % of land chart due to issue reading land area: {e}")
```
